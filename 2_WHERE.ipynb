# WHERE
조건에 맞는 데이터만 가져올 수 있습니다.  
`
SELEECT
    컬럼명  
FROM  
    테이블명  
WHERE  
    조건문  
`


연산자를 이용하여 조건문을 작성할 수 있습니다.

연산자
* [비교연산자](#비교-연산자)
* 논리연산자  

## 비교 연산자
|연산자|의미|
|---|---|
|=|일치하는 경우|
|!=|일치하지 않는 경우|
|<|해당 값 **미만**인 경우|
|<=|해당 값 **이하**인 경우|
|>|해당 값 **초과**인 경우|
|>=|해당 값 **이상**인 경우|

```python
import sqlite3
from prettytable import from_db_cursor

# SQLite 데이터베이스에 연결하기 위해 connect 생성
con = sqlite3.connect('./database/chinook.db')
con.row_factory = sqlite3.Row

# cursor 생성
cur = con.cursor()
```

```python
# customerid가 5 이하인 경우
cur.execute("""
    SELECT *
    FROM customers
    WHERE customerid <= 5
    """)

from_db_cursor(cur)
```

## 논리연산자

|연산자|의미|
|---|---|
|AND|두 식을 결합하여 두 식이 모두 참인경우 1 (True), 그렇지 않은 경우 0 (False)|
|OR|두 식을 결합하여 두 식 중 하나라도 참인경우 1 (True), 그렇지 않은 경우 0 (False)|
|NOT|해당 연산이 참이 아닌 경우 (ex. NOT IN, NOT EXISTS, NOT BETWEEN)|
|BETWEEN|지정한 범위 내의 값인 경우|
|IN|IN 뒤의 값에 포함되는 값인 경우|
|LIKE|패턴에 맞는 값인 경우|

```python
cur.execute("""
    SELECT * 
    FROM employees
    WHERE lastname = 'Adams' and firstname = 'Andrew'
    """)
from_db_cursor(cur)
```

### BETWEEN
지정한 범위 내의 값이면 참입니다.

```python
# BETWEEN a AND b -> a~b 범위 안이면 참 (a, b 포함)
cur.execute("""
    SELECT *
    FROM customers
    WHERE customerid BETWEEN 15 AND 20
    """)

from_db_cursor(cur)
```

### IN
지정한 목록에 있는 값 중 하나와 일치하면 참입니다.

```python
cur.execute("""
    SELECT *
    FROM customers
    WHERE customerid IN (1, 5, 10)
    """)

from_db_cursor(cur)
```

### LIKE
패턴과 일치하는 경우 참입니다.  
%와 _를 이용하여 패턴을 지정할 수 있습니다.  
%와 _는 모든 문자와 매칭되며, %는 문자열, _는 문자 하나를 의미합니다.

```python
# 전화번호가 +55로 시작하는 경우
cur.execute("""
    SELECT *
    FROM customers
    WHERE phone LIKE '+55%'
    """)

from_db_cursor(cur)
```

```python
# 전화번호에 00을 포함하는 경우
cur.execute("""
    SELECT *
    FROM customers
    WHERE phone LIKE '%00%'
    """)

from_db_cursor(cur)
```

```python
# 성이 on으로 끝나는 경우
cur.execute("""
    SELECT *
    FROM customers
    WHERE lastname LIKE '%on'
    """)

from_db_cursor(cur)
```

```python
# state가 B로 끝나는 2글자인 경우
cur.execute("""
    SELECT *
    FROM customers
    WHERE state LIKE '_B'
    """)

from_db_cursor(cur)
```

```python
# state가 S로 시작하는 두글자일 경우
cur.execute("""
    SELECT *
    FROM customers
    WHERE state LIKE 'S_'
    """)

from_db_cursor(cur)
```

```python
# email 도메인이 2글자인 경우 (. 뒤에 2글자인 경우) 5개만 값 가져오기
cur.execute("""
    SELECT *
    FROM customers
    WHERE email LIKE '%.__'
    LIMIT 5
    """)

from_db_cursor(cur)
```

```python
# datetime도 가능
cur.execute("""
    SELECT *
    FROM employees
    WHERE hiredate LIKE '2003-10-17%'
    """)

from_db_cursor(cur)
```

```python
# = 와 같이 사용할수도 있음
cur.execute("""
    SELECT *
    FROM customers
    WHERE state LIKE 'SP'
    """)

from_db_cursor(cur)
```
